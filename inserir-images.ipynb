```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from pywinauto.application import Application
import pyautogui
import time

caminho = r'inserirCaminho' #<<<<<Manter o r, barras \
numero_card = 'BC-2347'
board = '69' # 69 = Santander / 71 = Gooroo
email = 'emailparaacesso'
```

```python
driver = webdriver.Chrome()  #Iniciar chrome
```

```python
driver.get(f'linkPara{board}acessar{numero_card}')
driver.maximize_window()
WebDriverWait(driver, 30).until(
    EC.presence_of_element_located((By.CSS_SELECTOR, "#google-auth-button"))
)
driver.find_element(By.CSS_SELECTOR, "#google-auth-button").click()
driver.find_element(By.CSS_SELECTOR, "#identifierId").send_keys(f"{email}") 
driver.find_element(By.CSS_SELECTOR, "#identifierNext").click()

#>>>Terminar a conexão com Google
```

```python
#Quando logar, executar esse script

# driver.find_element(By.XPATH, "/html/body/div[25]/div[4]/div/div[2]/div/div/section/div/div/div[2]/div/div[1]/div[2]/div[2]/div/div[1]/div/span/span/div/div/div[2]/div/div[1]").click()
# time.sleep(5)

image_number = 1

while True:
    try:
                
        pyautogui.press('down')

        image_path = f'{caminho}\Screenshot_{image_number}.jpg'

        print("clicando para enviar imagem")

        upload_button = driver.find_element(By.XPATH, "//button[@aria-label='Adicionar imagem, vídeo ou arquivo']").click()

        time.sleep(3)

        pyautogui.press('down')


        print("Abrindo tela")

        app = Application().connect(title_re="Abrir", class_name="#32770")
        dlg = app.window(title_re="Abrir")

        print("Tela aberta")

        dlg.set_focus()  

        print("Inserindo nome de imagem")        


        dlg['Edit'].set_edit_text(image_path)

        print("Nome inserido")        

        dlg.wait('ready', timeout=10)

        print("subindo imagem")

        # Pressiona a tecla ENTER para confirmar o upload
        dlg.type_keys("{ENTER}")

        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "img[src*='Screenshot']"))
        )

        time.sleep(3)

        print("Imagem carregada com sucesso!")

        pyautogui.press('down')

        print(f"Imagem {image_number} carregada com sucesso. Iniciando processo")

        image_number += 1

        print(f"Novo processo iniciado com imagem {image_number}")

        time.sleep(5)

    except Exception as e:
        print(f"Não foi possível localizar ou clicar no item de índice {image_number}. Erro: {e}")
        break
```
